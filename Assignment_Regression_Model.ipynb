# Question 1 :

What is Simple Linear Regression (SLR)? Explain its purpose.

# Answer :

Simple Linear Regression (SLR) is a statistical method that models the relationship between one continuous predictor \(x\) and one continuous outcome \(y\) using a straight line.

The equation of the model is:

\[y = \beta_0 + \beta_1 x + \varepsilon\]

- \(y\) is the dependent variable (what you’re trying to predict).
- \(x\) is the independent variable (the predictor).
- \(\beta_0\) is the intercept (value of \(y\) when \(x=0\)).
- \(\beta_1\) is the slope (change in \(y\) for a one‑unit increase in \(x\)).
- \(\varepsilon\) is the random error term (captures everything \(x\) doesn’t explain).

Purpose.

1.Model the relationship – Quantify how \(y\) changes with \(x\) (e.g., how house price increases with area).
2. Predict – Use the fitted line to predict \(y\) for new \(x\) values.
3. Inference – Test if the slope \(\beta_1\) is significantly different from zero → does \(x\) actually influence \(y\)?
4. Interpretation – The coefficients tell you the direction and strength of the association.

Key assumptions- Linearity – The relationship is a straight line.
- Independence – Observations aren’t correlated.
- Homoscedasticity – Errors have constant variance.
- Normality of errors – Errors are normally distributed.
- No perfect multicollinearity (trivial here since there’s only one predictor).

When to use- You have one numeric predictor and a continuous target.
- You want a simple, interpretable model for prediction or inference.

Example: If you want to predict a person’s weight (\(y\)) based on their height (\(x\)), SLR fits a line weight = β0 + β1 × height and tells you the trend.

# Question 2:

What are the key assumptions of Simple Linear Regression?

# Answer:

Simple Linear Regression (SLR) relies on these key assumptions:

1. Linearity
The relationship between \(x\) and \(y\) is a straight line.
Check with: Scatterplot of \(x\) vs. \(y\) or residuals vs. fitted values.

2. Independence
Observations are not correlated with each other.
Common issue: Time‑series data (e.g., stock prices) where residuals might be autocorrelated.
Check with: Durbin‑Watson test.

3. Homoscedasticity
The variance of the residuals is constant across all levels of \(x\).
Check with: Residuals vs. fitted plot (look for a “funnel” shape).

4. Normality of errors
The residuals are normally distributed.
Check with: Q‑Q plot of residuals or Shapiro‑Wilk test.

5. No perfect multicollinearity
(Trivial in SLR since there’s only one predictor.)
The predictor \(x\) isn’t a constant and has some variation.


# Question 3:

 Write the mathematical equation for a simple linear regression model and
explain each term.

# Answer:

The mathematical equation for a simple linear regression model is:

\[y = \beta_0 + \beta_1 x + \varepsilon\]

Here's what each term means:

- \(y\): The dependent variable (also called the outcome, target, or response variable). This is what we're trying to predict.
- \(x\): The independent variable (also called the predictor, feature, or regressor). This is the input used to predict \(y\).
- \(\beta_0\): The intercept (or constant term). It's the expected value of \(y\) when \(x = 0\).
- \(\beta_1\): The slope coefficient. It represents the change in \(y\) for a one-unit increase in \(x\), holding all else constant.
- \(\varepsilon\): The error term (or residual). It captures the variability in \(y\) that's not explained by \(x\).

In short, the model says: "The outcome \(y\) is a linear function of \(x\), plus some random noise."

# Question 5:

 What is the method of least squares in linear regression?

# Answer:

The method of least squares is a way to estimate the coefficients (\(\beta_0\), \(\beta_1\)) in a linear regression model by minimizing the sum of squared residuals.

Mathematically, it minimizes this objective:

\[\sum_{i=1}^{n} (y_i - (\beta_0 + \beta_1 x_i))^2\]

- \(y_i\) is the actual value,
- \(\beta_0 + \beta_1 x_i\) is the predicted value,
- The difference \((y_i - \hat{y}_i)\) is a residual.

The solution gives the best‑fitting line that minimizes total squared error. This approach is called Ordinary Least Squares (OLS).

Key properties- Unbiased: Estimates are unbiased if assumptions hold.
- Closed‑form solution: The optimal \(\hat{\beta}\) has a formula involving \(X^T X\) and \(X^T y\).
- Sensitive to outliers: Squared errors penalize large residuals heavily.


# Question 6:

What is Logistic Regression? How does it differ from Linear
Regression?
# Answer:

Logistic Regression is a statistical model used for binary classification problems (predicting outcomes like 0/1, yes/no, spam/not spam).

Key differences from Linear Regression1. Output:
    - Linear Regression: Predicts a continuous value (e.g., house price).
    - Logistic Regression: Predicts a probability of belonging to a class (e.g., P(spam)).
2. Equation:
    - Linear: \(y = \beta_0 + \beta_1 x\)
    - Logistic: \(P(y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x)}}\) (sigmoid function squashes output to 0–1).
3. Purpose:
    - Linear: Predicts a quantity.
    - Logistic: Predicts a probability → used for classification (threshold e.g., 0.5).
4. Loss function:
    - Linear: Minimizes squared error.
    - Logistic: Maximizes likelihood (minimizes log loss).

When to use- Use Logistic Regression for binary outcomes (spam detection, disease diagnosis).
- Linear Regression for continuous outcomes (predicting prices, heights).

Logistic Regression is interpretable and works well for linearly separable data

# Question 7:

Name and briefly describe three common evaluation metrics for regression
models.
# Answer:

1. Mean Absolute Error (MAE)
    - Measures the average magnitude of errors between predicted (\(\hat{y}\)) and actual (\(y\)) values.
    - Formula: \(\text{MAE} = \frac{1}{n} \sum |y_i - \hat{y}_i|\)
    - Pros: Easy to interpret (same units as target), robust to outliers.
    - Cons: Doesn't penalize large errors heavily.

2. Mean Squared Error (MSE)
    - Measures the average of squared differences between predicted and actual values.
    - Formula: \(\text{MSE} = \frac{1}{n} \sum (y_i - \hat{y}_i)^2\)
    - Pros: Differentiable, penalizes large errors more (useful if big errors are costly).
    - Cons: Units are squared, sensitive to outliers.

3. R-squared (R²)
    - Measures proportion of variance in \(y\) explained by the model.
    - Formula: \(R^2 = 1 - \frac{\text{Sum of squared residuals}}{\text{Total sum of squares}}\)
    - Pros: Intuitive goodness-of-fit measure (0-1, higher is better).
    - Cons: Can be misleading with many predictors (use Adjusted R² instead).

# Question 8:

 What is the purpose of the R-squared metric in regression analysis?

# Answer:

R-squared (\(R^2\)) measures how much of the total variation in the dependent variable (\(y\)) is explained by the regression model.

In other words, it tells you the proportion of variance explained by the predictors.

\[R^2 = 1 - \frac{\text{Sum of squared residuals}}{\text{Total sum of squares}}\]

- \(R^2 = 1\): The model explains all the variation in \(y\) (perfect fit).
- \(R^2 = 0\): The model explains none of the variation (as good as using the mean).
- \(R^2\) is usually between 0 and 1 (can be negative for a really bad model).

Purpose- Goodness-of-fit: How well does the model fit the data?
- Model comparison: Compare \(R^2\) across nested models (but watch out for overfitting).
- Interpretation: E.g., “The model explains 80% of the variance in house prices.”

Limitations: \(R^2\) always increases with more predictors → use Adjusted \(R^2\) for model selection.

# Question 9:

Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.

(Include your Python code and output in the code box below.)

# Answer:

```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data
X = np.array([[1], [2], [3], [4], [5]])  # predictor
y = np.array([2, 3, 5, 7, 11])           # target

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")

# Optional: Plot the data and fit line
plt.scatter(X, y, label="Data")
plt.plot(X, model.predict(X), color='red', label="Fit line")
plt.legend()
plt.show()

```

- Magnitude of \(\beta_1\): How much \(y\) changes with a 1-unit \(x\) change.
- p-value of \(\beta_1\): Tests if the slope is significantly different from zero (i.e., \(x\) matters).
